```python
import pandas as pd
import sys
sys.path.append('..')
from src.data_loader import load_product_lookup

products = load_product_lookup()

#flag organic
products['is_organic'] = products['product_name'].str.contains(
    'Organic', case=False, na=False).astype(int)

print(f"Product lookup shape: {products.shape}")
print(f"Columns: {list(products.columns)}")
print(f"Organic products: {products['is_organic'].sum():,}")
print(f"Non-organic: {(products['is_organic']==0).sum():,}")
print()
print(products.head(5).to_string())

# Save
products.to_parquet('../data/processed/product_lookup.parquet', index=False)
print("Saved to data/processed/product_lookup.parquet")
```
